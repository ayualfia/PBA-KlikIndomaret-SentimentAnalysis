```python
# 03-amazon-shopping-reviews-data-preprocessing.ipynb
# Tujuan: Membersihkan dan mempreproses data review Amazon Shopping untuk analisis sentimen dan ekstraksi fitur.
# Data input: amazon_reviews.csv (dari file 02) yang berisi review berbahasa Inggris.

# Instalasi dan impor library
try:
    import pandas as pd
    import numpy as np
    import re
    import string
    import emoji
    import contractions
    from symspellpy import SymSpell, Verbosity
    from langdetect import detect, LangDetectException
    import pkg_resources
    import matplotlib.pyplot as plt
    import seaborn as sns
    from textblob import TextBlob
    from wordcloud import WordCloud
    import nltk
    import os
except ImportError as e:
    print(f"Error mengimpor library: {e}")
    !pip install pandas numpy emoji contractions symspellpy langdetect matplotlib seaborn textblob wordcloud nltk
    import pandas as pd
    import numpy as np
    import re
    import string
    import emoji
    import contractions
    from symspellpy import SymSpell, Verbosity
    from langdetect import detect, LangDetectException
    import pkg_resources
    import matplotlib.pyplot as plt
    import seaborn as sns
    from textblob import TextBlob
    from wordcloud import WordCloud
    import nltk
    import os

# Unduh resource NLTK yang diperlukan
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)
nltk.download('omw-1.4', quiet=True)

# Impor bagian spesifik dari NLTK
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from collections import Counter

# Mengatur gaya visualisasi
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (15, 10)

# Membaca data dari file CSV
file_path = 'amazon_reviews.csv'
if not os.path.exists(file_path):
    print(f"File {file_path} tidak ditemukan. Pastikan file ada di direktori yang sama atau ubah nama file.")
else:
    df_amazon_reviews = pd.read_csv(file_path)
    print("Informasi DataFrame awal:")
    df_amazon_reviews.info()

    # Tambahkan kolom ID incremental
    df_amazon_reviews['id'] = range(1, len(df_amazon_reviews) + 1)

    # Hapus kolom yang tidak dibutuhkan
    columns_to_drop = ["reviewId", "userName", "userImage", "replyContent", "repliedAt"]
    df_amazon_reviews = df_amazon_reviews.drop(columns=[col for col in columns_to_drop if col in df_amazon_reviews.columns])

    # Polaritas sentimen sebelum text preprocessing
    df_amazon_reviews['sentiment_polarity'] = df_amazon_reviews['content'].apply(
        lambda x: TextBlob(str(x)).sentiment.polarity
    )
    df_amazon_reviews['sentiment_subjectivity'] = df_amazon_reviews['content'].apply(
        lambda x: TextBlob(str(x)).sentiment.subjectivity
    )
    df_amazon_reviews['sentiment_rating'] = df_amazon_reviews['score'].apply(
        lambda x: 'Positive' if x > 3 else ('Neutral' if x == 3 else 'Negative')
    )

    plt.figure(figsize=(15, 10))
    sns.scatterplot(
        x=df_amazon_reviews['sentiment_polarity'],
        y=df_amazon_reviews['sentiment_subjectivity'],
        hue=df_amazon_reviews['sentiment_rating'],
        edgecolor='white',
        palette="pastel"
    )
    plt.title("Amazon Shopping Reviews Sentiment Analysis (Sebelum Preprocessing)", fontsize=20)
    plt.xlabel("Sentiment Polarity")
    plt.ylabel("Sentiment Subjectivity")
    plt.grid(True)
    plt.show()

    # Preprocessing data
    print("\nPreprocessing Data: Memahami tipe dan informasi data awal.")
    print("Kolom seperti 'reviewCreatedVersion' dan 'appVersion' dapat memiliki nilai null.")

    if 'reviewCreatedVersion' in df_amazon_reviews.columns:
        unique_versions = df_amazon_reviews['reviewCreatedVersion'].unique()
        print("Versi aplikasi yang unik dalam review:")
        print(unique_versions)

        num_unique_versions = df_amazon_reviews['reviewCreatedVersion'].nunique()
        print(f"\nJumlah versi aplikasi dalam review yang diambil adalah: {num_unique_versions}")

    # Menangani nilai null
    if 'reviewCreatedVersion' in df_amazon_reviews.columns:
        df_amazon_reviews['reviewCreatedVersion'] = df_amazon_reviews['reviewCreatedVersion'].fillna('Unknown')
    if 'appVersion' in df_amazon_reviews.columns:
        df_amazon_reviews['appVersion'] = df_amazon_reviews['appVersion'].fillna('Unknown')

    # Pastikan kolom content tidak memiliki NaN dan dikonversi ke string
    df_amazon_reviews['content'] = df_amazon_reviews['content'].astype(str).fillna("")

    # Menghapus review yang kosong
    df_amazon_reviews = df_amazon_reviews[df_amazon_reviews['content'].str.strip() != ""]

    # Validasi bahasa
    def detect_language(text):
        try:
            return detect(text)
        except LangDetectException:
            return 'unknown'

    df_amazon_reviews['language'] = df_amazon_reviews['content'].apply(detect_language)
    df_amazon_reviews = df_amazon_reviews[df_amazon_reviews['language'] == 'en']
    df_amazon_reviews = df_amazon_reviews.drop(columns=['language'])

    # Menghapus emoji dan karakter spesial
    def remove_emojis(text):
        emoji_pattern = re.compile("["
                                   u"\U0001F600-\U0001F64F"
                                   u"\U0001F300-\U0001F5FF"
                                   u"\U0001F680-\U0001F6FF"
                                   u"\U0001F1E0-\U0001F1FF"
                                   u"\U00002702-\U000027B0"
                                   u"\U000024C2-\U0001F251"
                                   u"\U0001f926-\U0001f937"
                                   u"\U00010000-\U0010ffff"
                                   u"\u2640-\u2642"
                                   u"\u2600-\u2B55"
                                   u"\u200d"
                                   u"\u23cf"
                                   u"\u23e9"
                                   u"\u231a"
                                   u"\ufe0f"
                                   u"\u3030"
                                   "]+", flags=re.UNICODE)
        return emoji_pattern.sub(r'', text)

    def clean_text(text):
        if not isinstance(text, str):
            return ""
        text = remove_emojis(text)
        text = re.sub(r",(?!\s)", ", ", text)
        text = text.encode('ascii', 'replace').decode('ascii')
        text = re.sub(r"(?i)(?:https?:\/\/)?(?:www\.)?(?:[a-zA-Z0-9-.]+)(?:\.[a-zA-Z]{2,6})(?:\/[^\s\r\n]*)?", "", text)
        text = text.strip()
        text = re.sub(r'\s+', ' ', text)
        return text

    df_amazon_reviews['content'] = df_amazon_reviews['content'].apply(clean_text)
    print("\nHasil awal setelah menghapus emoji dan karakter spesial:")
    print(df_amazon_reviews[['content']].head())

    # Menghapus review yang kosong lagi
    df_amazon_reviews = df_amazon_reviews[df_amazon_reviews['content'] != '']
    df_amazon_reviews = df_amazon_reviews[df_amazon_reviews['content'].str.strip() != '']
    df_amazon_reviews = df_amazon_reviews.dropna(subset=['content'])

    print("\nMemeriksa apakah masih ada review kosong:")
    print(df_amazon_reviews[df_amazon_reviews['content'].str.strip() == ''])

    # Lowercasing reviews
    df_amazon_reviews['content'] = df_amazon_reviews['content'].astype(str)
    df_amazon_reviews['content'] = df_amazon_reviews['content'].apply(lambda x: x.lower())
    print("\nHasil setelah lowercasing:")
    print(df_amazon_reviews[['content']].head())

    # Expand contractions
    def expand_contractions(text):
        return contractions.fix(text)

    df_amazon_reviews['content'] = df_amazon_reviews['content'].apply(expand_contractions)
    print("\nHasil setelah expand contractions:")
    print(df_amazon_reviews[['content']].head())

    # Spell correction
    sym_spell = SymSpell(max_dictionary_edit_distance=2, prefix_length=7)
    dictionary_path = pkg_resources.resource_filename("symspellpy", "frequency_dictionary_en_82_765.txt")
    bigram_path = pkg_resources.resource_filename("symspellpy", "frequency_bigramdictionary_en_243_342.txt")
    sym_spell.load_dictionary(dictionary_path, term_index=0, count_index=1)
    sym_spell.load_bigram_dictionary(bigram_path, term_index=0, count_index=2)

    def enhanced_correct_spelling(text):
        text = re.sub(r"[^a-zA-Z\s]", "", text)
        words = word_tokenize(text)
        corrected_words = [
            sym_spell.lookup(w, Verbosity.CLOSEST, max_edit_distance=2)[0].term
            if sym_spell.lookup(w, Verbosity.CLOSEST, max_edit_distance=2) else w
            for w in words
        ]
        corrected_text = " ".join(corrected_words)
        suggestions = sym_spell.lookup_compound(corrected_text, max_edit_distance=2)
        return suggestions[0].term if suggestions else corrected_text

    df_amazon_reviews['final_cleaned'] = df_amazon_reviews['content'].apply(enhanced_correct_spelling)
    print("\nHasil setelah spell correction:")
    print(df_amazon_reviews[['final_cleaned']].head())

    # Handle negation
    def handle_negation(text):
        negation_patterns = r"\b(?:not|no|never|none|n't)\s+(\w+)"
        return re.sub(negation_patterns, lambda match: match.group(0).replace(" ", "_"), text)

    df_amazon_reviews['final_cleaned'] = df_amazon_reviews['final_cleaned'].apply(handle_negation)

    # Punctuation removal
    def remove_punctuation(text):
        return text.translate(str.maketrans("", "", string.punctuation))

    df_amazon_reviews['final_cleaned'] = df_amazon_reviews['final_cleaned'].apply(remove_punctuation)
    print("\nHasil setelah punctuation removal:")
    print(df_amazon_reviews[['final_cleaned']].head())

    # Tokenization
    df_amazon_reviews['token'] = df_amazon_reviews['final_cleaned'].apply(word_tokenize)
    print("\nHasil setelah tokenization:")
    print(df_amazon_reviews[['final_cleaned', 'token']].head())

    # Most frequent words sebelum penambahan stopwords tambahan
    all_words_before = [word for tokens in df_amazon_reviews['token'] for word in tokens if word.isalpha()]
    most_common_before = Counter(all_words_before).most_common(20)

    print("\n20 kata paling sering muncul sebelum stopwords removal:")
    print(most_common_before)

    # Stopwords removal
    english_stopwords = set(stopwords.words('english'))
    negation_words = {'not', 'no', 'never', "n't", 'none'}
    english_stopwords = english_stopwords.difference(negation_words)

    # Stopwords tambahan umum
    additional_stopwords = {
        'work', 'works', 'working', 'update', 'back', 'really',
        'much', 'still', 'time'
    }

    # Kata-kata dari most frequent words yang dianggap kurang bermakna
    frequent_unimportant_words = {
       'amazon', 'app', 'shopping', 'product', 'products', 'item', 'items',
       'order', 'orders', 'delivery', 'buy', 'purchase', 'cart',
       'get', 'use', 'like', 'would', 'one',
       'customer', 'account', 'day', 'days',
        'always', 'even', 'anything', 'everything', 'something', 'find'
    }

    additional_stopwords = additional_stopwords.union(frequent_unimportant_words)
    english_stopwords.update(additional_stopwords)

    def stopwords_removal(tokens):
        cleaned_tokens = [word for word in tokens if word.lower() not in english_stopwords and word.isalpha()]
        return [word for word in cleaned_tokens if '_' in word or word.lower() not in english_stopwords]

    df_amazon_reviews['token_after_sr'] = df_amazon_reviews['token'].apply(stopwords_removal)
    print("\nHasil setelah stopwords removal:")
    print(df_amazon_reviews[['token', 'token_after_sr']].head())

    # Visualisasi kata paling sering setelah stopwords removal
    def freq_words(tokens_column, top_n=20):
        all_words = [word for tokens in tokens_column for word in tokens]
        word_freq = Counter(all_words).most_common(top_n)
        if len(word_freq) == 0:
            print("Tidak ada kata untuk divisualisasikan.")
            return
        words, counts = zip(*word_freq)
        plt.figure(figsize=(10, 5))
        plt.bar(words, counts)
        plt.xticks(rotation=45)
        plt.xlabel("Words")
        plt.ylabel("Frequency")
        plt.title("Top Frequent Words")
        plt.show()

    freq_words(df_amazon_reviews['token_after_sr'])

    # Lemmatization
    lemmatizer = WordNetLemmatizer()

    def lemmatize_text(tokens):
        return [lemmatizer.lemmatize(token) for token in tokens]

    df_amazon_reviews['lemmatized_tokens'] = df_amazon_reviews['token_after_sr'].apply(lemmatize_text)
    print("\nHasil setelah lemmatization:")
    print(df_amazon_reviews[['final_cleaned', 'token_after_sr', 'lemmatized_tokens']].head())

    # Gabungkan kembali token menjadi kalimat utuh
    df_amazon_reviews['cleaned_text'] = df_amazon_reviews['lemmatized_tokens'].apply(lambda tokens: ' '.join(tokens))
    print("\nHasil setelah menggabungkan token:")
    print(df_amazon_reviews[['lemmatized_tokens', 'cleaned_text']].head())

    # Sentiment analysis setelah preprocessing
    df_amazon_reviews['sentiment_polarity'] = df_amazon_reviews['cleaned_text'].apply(
        lambda x: TextBlob(str(x)).sentiment.polarity
    )
    df_amazon_reviews['sentiment_subjectivity'] = df_amazon_reviews['cleaned_text'].apply(
        lambda x: TextBlob(str(x)).sentiment.subjectivity
    )

    plt.figure(figsize=(15, 10))
    sns.scatterplot(
        x=df_amazon_reviews['sentiment_polarity'],
        y=df_amazon_reviews['sentiment_subjectivity'],
        hue=df_amazon_reviews['sentiment_rating'],
        edgecolor='white',
        palette="pastel"
    )
    plt.title("Amazon Shopping Reviews Sentiment Analysis (Setelah Preprocessing)", fontsize=20)
    plt.xlabel("Sentiment Polarity")
    plt.ylabel("Sentiment Subjectivity")
    plt.grid(True)
    plt.show()

    # WordCloud
    text_for_wordcloud = ' '.join(df_amazon_reviews['cleaned_text'].astype(str))
    if text_for_wordcloud.strip():
        wordcloud_after = WordCloud(width=800, height=400, background_color='white').generate(text_for_wordcloud)
        plt.figure(figsize=(12, 6))
        plt.imshow(wordcloud_after, interpolation='bilinear')
        plt.axis('off')
        plt.title("WordCloud dari Amazon Shopping Reviews (Setelah Preprocessing)", fontsize=16)
        plt.show()

    # Ekstrak ke CSV baru
    df_amazon_reviews.to_csv('df_amazon_reviews2.csv', index=False, encoding='utf-8')
    print("\nDataFrame telah disimpan ke df_amazon_reviews2.csv")
```
