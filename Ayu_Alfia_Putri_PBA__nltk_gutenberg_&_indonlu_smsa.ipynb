## nltk-gutenberg

```python
import nltk
from nltk.tokenize import word_tokenize

# Download NLTK data
nltk.download('gutenberg')
nltk.download('punkt')

# Load sample text from Gutenberg corpus
text = nltk.corpus.gutenberg.raw('austen-emma.txt')[:1000]  # First 1000 chars of Emma

# Tokenize the text
tokens = word_tokenize(text)

# Print first 20 tokens
print("First 20 tokens:", tokens[:20])

# Basic statistics
print("Total tokens:", len(tokens))
print("Unique tokens:", len(set(tokens)))
```

```python
from collections import Counter

word_freq = Counter([t.lower() for t in tokens]).most_common(10)
print("Top 10 tokens:", word_freq)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
import pandas as pd

columns_name = ['text', 'label']
df_train = pd.read_csv('data/raw/train_preprocess.tsv', sep='\t', header=None, names=columns_name)
df_train.head()
```

```python
sample_texts = df_train['text'][:3].tolist()
labels = df_train['label'][:3].tolist()

sample_texts, labels
```

```python
stemmer = StemmerFactory().create_stemmer()

for i, text in enumerate(sample_texts):
    print(f"\nContoh Teks {i+1} (Label: {labels[i]}):")
    print("Teks Asli:", text)

    tokens = word_tokenize(text)
    print("Token:", tokens)
    print("Jumlah Token:", len(tokens))
    print("Token Unik:", len(set(tokens)))

    stemmed_text = stemmer.stem(text)
    stemmed_tokens = word_tokenize(stemmed_text)
    print("Token setelah Stemming:", stemmed_tokens)
    print("Jumlah Token setelah Stemming:", len(stemmed_tokens))

    with open(f"data/processed/tokens_sample_{i+1}.txt", "w", encoding="utf-8") as f:
        f.write("\n".join(tokens))
    print(f"Token disimpan ke data/processed/tokens_sample_{i+1}.txt")
```

```python
from collections import Counter

all_tokens = []
for text in sample_texts:
    all_tokens.extend(word_tokenize(text.lower()))

all_tokens[:30]  # cek contoh token (opsional)
```

```python
word_freq = Counter(all_tokens).most_common(10)
print("10 Kata Teratas:", word_freq)
```
